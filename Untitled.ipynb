```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_stata("FDI.dta")

# Danh sách các quốc gia
countries = df['ID'].unique()
n_countries = len(countries)

# Chọn màu cho từng quốc gia (sử dụng palette của Seaborn)
palette = sns.color_palette("tab20", n_colors=n_countries)

# Xác định số hàng và cột cho subplot
n_cols = 3
n_rows = (n_countries + n_cols - 1) // n_cols

# Tạo figure với nhiều subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows), sharex=True, sharey=True)
axes = axes.flatten()  # chuyển thành mảng 1 chiều để dễ lặp

for i, country in enumerate(countries):
    df_country = df[df['ID'] == country]
    sns.lineplot(
        ax=axes[i], 
        data=df_country, 
        x="TIME", y="FDI", 
        marker="o", 
        color=palette[i],  # màu khác nhau
        linewidth=2.0
    )
    axes[i].set_title(country, fontsize=12, fontweight='bold')
    axes[i].grid(True, linestyle='--', alpha=0.5)

# Ẩn các subplot thừa nếu có
for j in range(i+1, len(axes)):
    axes[j].axis('off')

# Nhãn chung
fig.text(0.5, 0.04, 'Năm', ha='center', fontsize=14, fontweight='bold')
fig.text(0.04, 0.5, 'FDI', va='center', rotation='vertical', fontsize=14, fontweight='bold')
fig.suptitle("FDI theo thời gian cho các quốc gia", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0.03, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_stata("FDI.dta")

# Lọc khoảng thời gian 2000-2024
df_period = df[(df['TIME'] >= 2000) & (df['TIME'] <= 2024)]

# Tính trung bình TAX2 theo quốc gia (%GDP)
mean_tax2 = df_period.groupby('ID')['TAX2'].mean().reset_index()
mean_tax2 = mean_tax2.sort_values(by='TAX2', ascending=False)  # Sắp xếp từ cao xuống thấp

# Tạo gradient màu theo giá trị TAX2
norm = plt.Normalize(mean_tax2['TAX2'].min(), mean_tax2['TAX2'].max())
colors = plt.cm.Blues(norm(mean_tax2['TAX2']))  # gradient xanh nhạt -> đậm

# Vẽ dot plot
plt.figure(figsize=(12,6))
plt.scatter(mean_tax2['ID'], mean_tax2['TAX2'], s=150, c=colors)

# Nâng cấp hiển thị
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.title("TAX2 trung bình theo quốc gia (2000-2024, %GDP)", fontsize=16, fontweight='bold')
plt.ylabel("TAX2 trung bình (%GDP)", fontsize=12)
plt.xlabel("Quốc gia", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Thêm colorbar để thấy độ đậm nhạt tương ứng với giá trị
sm = plt.cm.ScalarMappable(cmap="Blues", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label('TAX2 trung bình (%GDP)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Đọc dữ liệu
df = pd.read_stata("FDI.dta")

# Lọc dữ liệu từ 2000-2024
df_period = df[(df['TIME'] >= 2000) & (df['TIME'] <= 2024)]

# Tạo biểu đồ động TAX2 theo thời gian với màu cố định theo quốc gia
fig = px.scatter(
    df_period,
    x="ID",                # trục X: quốc gia
    y="TAX2",              # trục Y: TAX2 (%GDP)
    animation_frame="TIME",# frame theo năm
    size="TAX2",           # kích thước chấm theo giá trị TAX2
    color="ID",            # màu cố định cho mỗi quốc gia
    title="TAX theo quốc gia (2000-2024, %GDP)",
    labels={"TAX2":"TAX (%GDP)", "ID":"Quốc gia"},
    size_max=25            # max size chấm
)

# Tùy chỉnh layout
fig.update_layout(
    xaxis_tickangle=-45,
    xaxis_title="Quốc gia",
    yaxis_title="TAX (%GDP)",
    legend_title="Quốc gia"
)

fig.show()
```
