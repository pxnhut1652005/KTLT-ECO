```python
import pandas as pd

df = pd.read_stata("FIN.dta")
# Cài đặt hiển thị
pd.set_option("display.max_columns", None)  # Hiện tất cả cột
pd.set_option("display.width", 1000)        # Mở rộng chiều rộng in (có thể tăng lên 2000 nếu còn hẹp)
pd.set_option("display.max_colwidth", None) # Không giới hạn độ rộng ô
df = df.rename(columns={"FI": "FE"})
df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
variables = ["Ln_Y", "IF", "FD", "FE"]

for var in variables:
    plt.figure(figsize=(10,5))
    sns.lineplot(data=df, x="Time", y=var)
    plt.title(f"Trend of {var} over time")
    plt.xlabel("Time")
    plt.ylabel(var)
    plt.grid(True)
    plt.tight_layout()
 

```

```python
for var in variables:
    if var == "Ln_Y":
        continue
    plt.figure(figsize=(7,5))
    sns.regplot(data=df, x=var, y="Ln_Y", scatter_kws={"alpha":0.5})
    plt.title(f"Relationship between {var} and Ln_Y")
    plt.xlabel(var)
    plt.ylabel("Ln_Y")
    plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import os

# Tạo thư mục lưu hình
os.makedirs("dual_axis_plots_pooled", exist_ok=True)

pairs = [
    ("Ln_Y", "IF"),
    ("Ln_Y", "FD"),
    ("Ln_Y", "FE"),
]

# Lấy giá trị trung bình theo năm toàn bộ quốc gia (CHỈ CỘT SỐ)
df_avg = df.groupby("Time").mean(numeric_only=True).reset_index()

for v1, v2 in pairs:
    fig, ax1 = plt.subplots(figsize=(10,5))

    # Trục Y thứ nhất
    ax1.set_xlabel("Time")
    ax1.set_ylabel(v1, color="tab:blue")
    ax1.plot(df_avg["Time"], df_avg[v1], color="tab:blue", linewidth=2)
    ax1.tick_params(axis="y", labelcolor="tab:blue")

    # Trục Y thứ hai
    ax2 = ax1.twinx()
    ax2.set_ylabel(v2, color="tab:red")
    ax2.plot(df_avg["Time"], df_avg[v2], color="tab:red", linewidth=2, linestyle="--")
    ax2.tick_params(axis="y", labelcolor="tab:red")

    # Tiêu đề
    plt.title(f"Dual-Axis Time Plot: {v1} vs {v2}")
    fig.tight_layout()

```
